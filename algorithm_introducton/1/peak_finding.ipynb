# Peak finding
- compare 2 algorithms
- random array and sort array

```python
import copy
import time

import numpy as np
import matplotlib.pyplot as plt
from scipy.interpolate import interp1d
```

```python
def brute_force(A):
    for i in range(1, len(A)-1):
        if A[i] > A[i-1] and A[i] > A[i+1]:
            print('FIND PEAK (BRUTE ALGORITHM)')
            return A[i]
        
    print('NOT FOUND (BRUTE ALGORITHM)')
    return None
    
def devide_conquer(A, i, j):
    n = (i + j)//2 
    if n < 0:
        print('NOT FOUND (DEVIDE CONQUER)')
        return None
    
    if A[n] > A[n-1] and A[n] > A[n+1]:
        print('FIND PEAK (DEVIDE CONQUER)')
        return A[n]
    elif A[n] <= A[n-1]:
        return devide_conquer(A, i, n-1)
    elif A[n] <= A[n+1]:
        return devide_conquer(A, n+1, j)
```

```python
count = 5
sizes = np.logspace(start=1, stop=count, num=count, base=10)
brute_force_times = []
devide_conquer_times = []


for s in sizes:
    arr = list(np.linspace(1, s, num=s))
    arr.append(float('-inf'))
    arr.insert(0, float('-inf'))

    start = time.time()
    brute_force(copy.copy(arr))
    end = time.time()
    brute_force_times.append(end - start)
    
    start = time.time()
    devide_conquer(copy.copy(arr), 0, len(arr))
    end = time.time()
    devide_conquer_times.append(end - start)
```

```python
# Spline https://docs.scipy.org/doc/scipy/reference/generated/scipy.interpolate.interp1d.html
sp_brute = interp1d(sizes, brute_force_times)
sp_devide = interp1d(sizes, devide_conquer_times)
sp_x = np.linspace(sizes[0], sizes[-1], num=100)

brute_label,  = plt.plot(sizes, brute_force_times, 'ro', label='brute_force')
devide_label,  = plt.plot(sizes, devide_conquer_times, 'bo', label='devide_conquer')
plt.plot(sp_x, sp_brute(sp_x), 'r-',
         sp_x, sp_devide(sp_x), 'b-')

plt.legend(handles=[brute_label, devide_label])
plt.xscale('log')
plt.xlabel('array length')
plt.ylabel('computational time[sec]')
plt.show()
```
